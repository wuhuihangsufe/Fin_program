```python
import pandas as pd
import os
import numpy as np
import matplotlib.pyplot as plt
import time
from pandas.tseries.offsets import *
from scipy import stats
import statsmodels.api as sm
import datetime
import math
```

```python
#1.1导入IPO数据
program_path = 'D:\\China A share\\'
Start_Date = '1997-01-01'
End_Date = '2019-06-30'
save_list = ['permno', 'ShortName', 'ListedDate', 'DelistedDate',
             'IPOIndustryCode', 'IPOIndustryName', 'LatestIndustryName',
             'LatestIndustryCode']

IPO = pd.read_excel(program_path + 'rawwind\\ER_IPO.xlsx', encoding='utf-16')
IPO = IPO.iloc[2:,:]
IPO['ListedDate'] = pd.to_datetime(IPO['ListedDate'], format='%Y/%m/%d')
IPO['DelistedDate'] = pd.to_datetime(IPO['DelistedDate'], format='%Y/%m/%d')
IPO['permno'] = IPO['Symbol'].copy()
IPO['Symbol'] = IPO['Symbol'].apply(lambda x: x[:3])
# 保留A股
IPO = IPO[
    (IPO['permno'].str.startswith('0')) | (IPO['permno'].str.startswith('3')) | (IPO['permno'].str.startswith('6'))]
# 剔除 688开头的科创板
IPO = IPO[IPO['Symbol'] != '688']

# 调整和剔除代码变更
IPO['permno'].replace('000022', '001872', inplace=True)
IPO['permno'].replace('601313', '601360', inplace=True)
IPO['permno'].replace('600087', '601975', inplace=True)
IPO.drop(index=IPO[IPO['permno'] == '600849'].index, inplace=True)

#为了IPO数据重复，发生二次上市的企业都用不同代码标注
# debug for 600018
temp_index = IPO[(IPO['permno'] == '600018') & (IPO['ListedDate'] == pd.datetime(2000, 7, 19))].index
IPO.loc[temp_index, 'permno'] = 'T00018'
#debug for 601975
temp_index = IPO[(IPO['permno'] == '601975') & (IPO['ListedDate'] == pd.datetime(1997, 6, 12))].index
IPO.loc[temp_index, 'permno'] = '600087'


IPO_sub = IPO.loc[:, save_list]

IPO_sub.to_pickle(program_path + 'temp\\IPO_list.pickle')
IPO_sub.to_csv(program_path + 'temp\\IPO_list.csv')
```

```python
#1.2生成完美的list
#生成股票list和1997-2018每季度最后一天的时间序列表格
qlist=pd.DataFrame()
for k in range(2019-1997+1):
    qlist=np.append(qlist,pd.datetime(1997+k,3,31))
    qlist=np.append(qlist,pd.datetime(1997+k,6,30))
    qlist=np.append(qlist,pd.datetime(1997+k,9,30))
    qlist=np.append(qlist,pd.datetime(1997+k,12,31))
#concad，得到最长序列，根据IPO裁剪
#得到完美list，用reportdate汇整
qlist=qlist[qlist<pd.datetime(2019,8,1)]
```

```python
permnolist=[]
for k in IPO_sub['permno'].values:
    q=pd.Series(qlist)
    permnolist.append(q)
q=pd.concat(permnolist,keys=IPO_sub['permno'].values).reset_index().drop('level_1',axis=1,inplace=False).rename(columns={'level_0':'permno',0:'REPORT_PERIOD'},inplace=False)
q=pd.merge(q,IPO_sub,on='permno',how='left')
q=q[(q['REPORT_PERIOD']>=q['ListedDate'])&((q['REPORT_PERIOD']<=q['DelistedDate'])|(q['DelistedDate'].isnull()))].loc[:,['permno','REPORT_PERIOD','ListedDate','DelistedDate']]
```

```python
#在2002年以前只保留半年度报表
q=q[(q['REPORT_PERIOD']>pd.datetime(2002,1,1))|((q['REPORT_PERIOD'].map(lambda x:x.month==6))|(q['REPORT_PERIOD'].map(lambda x:x.month==12)))]
```

```python
#2.1 导入损益表
program_path = 'D:\\China A share\\'
income=pd.read_csv(program_path + 'rawwind\\ashareincome.csv')
income['permno']=income['S_INFO_WINDCODE'].str.split('.',expand=True)[0]

#转换日期格式
income['REPORT_PERIOD']=pd.to_datetime(income['REPORT_PERIOD'],format='%Y%m%d')
income['ANN_DT']=pd.to_datetime(income['ANN_DT'],format='%Y%m%d')
#保留A股数据
income=income[(income['permno'].str.startswith('0'))|(income['permno'].str.startswith('3'))|(income['permno'].str.startswith('60'))]
income.loc[(income['permno']=='601975')&(income['REPORT_PERIOD']<pd.datetime(2018,12,31)),'permno']='600087'
income.loc[(income['permno']=='600018')&(income['REPORT_PERIOD']<pd.datetime(2006,10,26)),'permno']='T00018'

#在存在更正前和更正后的数据情况下，保留未更正的数据
income=income.loc[(income['STATEMENT_TYPE']==408005000)|(income['STATEMENT_TYPE']==408001000)].sort_values(['permno','REPORT_PERIOD','STATEMENT_TYPE'])
income.drop_duplicates(subset=['permno','REPORT_PERIOD'],keep='last',inplace=True)
```

```python
#3.1 生成单季度损益表
lastqincome=qincome.drop(['OBJECT_ID',
       'S_INFO_WINDCODE', 'WIND_CODE', 'ANN_DT', 'STATEMENT_TYPE','MEMO',
       'CRNCY_CODE','ACTUAL_ANN_DT','S_INFO_COMPCODE','IS_CALCULATION', 'OPDATE', 'OPMODE'],axis=1)
lastqincome.loc[lastqincome['REPORT_PERIOD']>=pd.datetime(2001,12,31),'reportdate']=lastqincome['REPORT_PERIOD']+MonthEnd(3)
lastqincome.loc[lastqincome['REPORT_PERIOD']<pd.datetime(2001,12,31),'reportdate']=lastqincome['REPORT_PERIOD']+MonthEnd(6)
lastqincome.rename(columns={'REPORT_PERIOD':'lastrepodate'},inplace=True)
lastqincome.rename(columns={'reportdate':'REPORT_PERIOD'},inplace=True)
```

```python
qincomecol=qincome.columns.drop(['permno','REPORT_PERIOD','ListedDate','DelistedDate','OBJECT_ID',
       'S_INFO_WINDCODE', 'WIND_CODE', 'ANN_DT', 'STATEMENT_TYPE','MEMO',
       'CRNCY_CODE','ACTUAL_ANN_DT','S_INFO_COMPCODE','IS_CALCULATION', 'OPDATE', 'OPMODE'])
for k in qincomecol:
    lastqincome.rename(columns={k:k+'last'},inplace=True)
diffqincome=pd.merge(qincome,lastqincome,on=['permno','REPORT_PERIOD','ListedDate','DelistedDate'],how='left')
for k in qincomecol:
    diffqincome[k+'_diff']=diffqincome[k]-diffqincome[k+'last']
    diffqincome.drop([k,k+'last'],axis=1,inplace=True)
```

```python
#缺失数据作图
totalcount=diffqincome.groupby(['REPORT_PERIOD'])['permno'].count().reset_index().rename(columns={'permno':'TOTAL'},inplace=False)
missingcount=diffqincome[(diffqincome['ANN_DT'].isnull())].groupby(['REPORT_PERIOD'])['permno'].count().reset_index().rename(columns={'permno':'Missing'},inplace=False)
missingdel=diffqincome[(diffqincome['ANN_DT'].isnull())&((diffqincome['REPORT_PERIOD']+MonthEnd(4)<=diffqincome['DelistedDate'])|diffqincome['DelistedDate'].isnull())].groupby(['REPORT_PERIOD'])['permno'].count().reset_index().rename(columns={'permno':'Missing nonDEL'},inplace=False)
diffqincome_count=pd.merge(totalcount,missingcount,on=['REPORT_PERIOD'],how='left')
diffqincome_count=pd.merge(diffqincome_count,missingdel,on=['REPORT_PERIOD'],how='left')
diffqincome_count.to_csv(program_path + 'tempwind\\diffqincome_count.csv')
```
